```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load datasets from Excel file in Google Colab
from google.colab import files
uploaded = files.upload()
```

```python
# Replace 'Assignment.xlsx' with the uploaded filename
datasets = pd.ExcelFile("Assignment.xlsx")
```

```python
# Load individual sheets
userdetails = pd.read_excel(datasets, sheet_name='UserDetails.csv')
cookingsessions = pd.read_excel(datasets, sheet_name='CookingSessions.csv')
orderdetails = pd.read_excel(datasets, sheet_name='OrderDetails.csv')

```

```python
# ------------------- 1. DATA CLEANING ------------------- #
print("Initial Data Overview:")
print(userdetails.info())
print(cookingsessions.info())
print(orderdetails.info())

# Clean UserDetails dataset
def clean_userdetails(df):
    df = df.drop(columns=['Phone', 'Email'], errors='ignore')  # Drop unnecessary columns
    df = df.dropna().reset_index(drop=True)  # Remove rows with missing values
    df['Registration Date'] = pd.to_datetime(df['Registration Date'])  # Convert date column
    return df

# Clean CookingSessions dataset
def clean_cookingsessions(df):
    df = df.dropna().reset_index(drop=True)  # Remove missing values
    df['Session Start'] = pd.to_datetime(df['Session Start'])
    df['Session End'] = pd.to_datetime(df['Session End'])
    return df

# Clean OrderDetails dataset
def clean_orderdetails(df):
    df = df.dropna().reset_index(drop=True)  # Remove missing values
    df['Order Date'] = pd.to_datetime(df['Order Date'])
    return df

# Apply cleaning functions
userdetails_clean = clean_userdetails(userdetails)
cookingsessions_clean = clean_cookingsessions(cookingsessions)
orderdetails_clean = clean_orderdetails(orderdetails)

```

```python
print("\nCleaned Data Overview:")
print("UserDetails:")
print(userdetails_clean.head())
print("CookingSessions:")
print(cookingsessions_clean.head())
print("OrderDetails:")
print(orderdetails_clean.head())
```

```python

import matplotlib.pyplot as plt

# Customer Age Distribution
plt.figure(figsize=(12, 6))
plt.hist(merged_data['Age'], bins=20, color='dodgerblue', edgecolor='black', alpha=0.7)
plt.title('Age Distribution of Customers', fontsize=16, fontweight='bold', color='darkblue')
plt.xlabel('Age', fontsize=14, fontweight='bold', color='darkblue')
plt.ylabel('Number of Customers', fontsize=14, fontweight='bold', color='darkblue')
plt.grid(True, linestyle='--', alpha=0.6)

# Adding custom ticks for the x-axis to make it more readable
plt.xticks(range(int(min(merged_data['Age'])), int(max(merged_data['Age'])) + 1, 2), fontsize=12)
plt.yticks(fontsize=12)

# Show plot
plt.tight_layout()
plt.show()

# Customer Location Distribution
plt.figure(figsize=(12, 6))
ax = merged_data['Location'].value_counts().plot(kind='bar', color='lightcoral', edgecolor='black', alpha=0.8)
plt.title('Customer Location Distribution', fontsize=16, fontweight='bold', color='darkred')
plt.xlabel('Location', fontsize=14, fontweight='bold', color='darkred')
plt.ylabel('Number of Customers', fontsize=14, fontweight='bold', color='darkred')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# Adding data labels to the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', fontweight='bold')

plt.grid(True, linestyle='--', alpha=0.6)

# Show plot
plt.tight_layout()
plt.show()
```

```python
# Order Frequency by Date
merged_data['Order Date'] = pd.to_datetime(merged_data['Order Date'])
orders_by_date = merged_data.groupby('Order Date').size()

plt.figure(figsize=(12, 6))
orders_by_date.plot(kind='line', color='purple')
plt.title('Order Frequency Over Time')
plt.xlabel('Date')
plt.ylabel('Number of Orders')
plt.grid(True)
plt.show()
```
